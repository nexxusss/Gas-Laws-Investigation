```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Helium

```python
df = pd.read_csv("boyleslaw_data_helium.csv")
df["1/V"] = 1/df["V"]
tab = df.style.set_caption("Data for helium")
```

```python
tab
```

```python
pressure = df["P"]
volume = df["V"]
inverse_vol = df["1/V"]

plt.scatter(volume, pressure)
plt.title("Relation between P and V for Helium")
plt.xlabel("Volume V(L)")
plt.ylabel("Pressure P(atm)")
```

```python
plt.scatter(inverse_vol, pressure)
plt.title("Relation between P and 1/V for Helium")
plt.xlabel("1/V(L^(-1))")
plt.ylabel("Pressure P(atm)")

#adding the trend line
z = np.polyfit(inverse_vol, pressure, 1)
p = np.poly1d(z)
plt.plot(inverse_vol, p(inverse_vol), "r-o")
plt.gcf().set_size_inches((5, 5))  
plt.show()
```

```python
#finding the slope of the linear relation between P and 1/V

x = np.array(inverse_vol).reshape(-1, 1)
y = np.array(pressure)

lr = LinearRegression()
lr.fit(x, y)

coeff = lr.coef_
inter = lr.intercept_
# used for the trend line with y = coeff*x + inter
print(f"Slope: {coeff}")
#print(f"Intercept: {inter}") to be ignored
print(f"equation fo the line: y = {coeff[0] - coeff[0]%0.01}x + ({inter - inter%0.01})")
print(f"R^2: {r2_score(y, coeff*x)}")

```

#### From the table and the data set, the experiment design sets n(mol) and the temperature T(K) as constants. From the ideal gas equation $ PV=nRT $ and the equation $ P = a\frac{1}{V}$ with $a$ the slope found previously, we deduce that the gas constant $R = \frac{a}{nT}$

```python
n = 0.05
T = 298
r = coeff[0]/(n*T)

print("The gas constant is R = " + str(r) + "L.atm/mol.K")
```

We indeed find that the gas constant $R = 0.0820  \hspace{0.1cm} L.atm/mol.K$
